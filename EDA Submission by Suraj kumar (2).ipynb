### Data Wrangling Code

````python
#replace null value to 0 in country col
#is used to fill missing values
df['children']=df['children'].fillna(0)
df['children'].isnull().sum()
````

````python
#replace null value in "agent" colum by mean
#This approach is useful for numerical columns where replacing missing values with the mean helps maintain the overall statistical properties of the dataset.
df['agent']=df['agent'].fillna(df['agent'].mean())
df['agent'].isnull().sum()
````

````python
#remove missing data of 0.40% in country
#is used to remove rows from the DataFrame
#dropping rows containing NaN values specifically in the 'country' column.
df=df.dropna(subset=['country'])
````

````python
#drop compant colum because its have 94% null value
#is used to drop (remove) the 'company' column
df=df.drop('company',axis=1)
````

````python
#now cheak shape and null value
print(df.shape)
print(df.isnull().sum())
````

````python
#info
df.info()
````

````python
#total stay in night
df['total_stay_night']=df['stays_in_weekend_nights']+df['stays_in_week_nights']
df['total_stay_night']
````

````python
#create colum revnue using total stay *adr
df['revenue']=df['total_stay_night']*df['adr']
df['revenue']
````

````python
#total guest come for each booking
df['total_guest']=df['adults']+df['children']+df['babies']
df['total_guest'].sum()
````

````python
#replace the value from colum is_cancelled 0 or 1 to not cancelled or is cancelled
df['is_cancelled']=df['is_canceled'].replace({0:'Not Cancelled',1:'Cancelled'})
df['is_cancelled']
````

````python
# do the same with reapated guest colum
df['is_repeated_guest']=df['is_repeated_guest'].replace({0:'Not Repeated',1:'Repeated'})
df['is_repeated_guest']
````

````python
#cheak revnue hotal wise
df.groupby('hotel')['revenue'].sum()
````

````python
# Calculate number of bookings for each hotel type
# Calculate percentages
city_hotel_bookings = df[df['hotel'] == 'City Hotel'].shape[0]
resort_hotel_bookings = df[df['hotel'] == 'Resort Hotel'].shape[0]

total_bookings = city_hotel_bookings + resort_hotel_bookings

city_hotel_percentage = (city_hotel_bookings / total_bookings) * 100
resort_hotel_percentage = (resort_hotel_bookings / total_bookings) * 100

print(f"Percentage of City Hotel bookings: {city_hotel_percentage:.2f}%")
print(f"Percentage of Resort Hotel bookings: {resort_hotel_percentage:.2f}%")
````
